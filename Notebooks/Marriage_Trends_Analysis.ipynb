```python
plt.figure(figsize=(10,6))
plt.hist(
    df["Age_at_Marriage"],
    bins=10,
    color="orange",
    edgecolor="white"
)

plt.xlabel("Age at Marriage")
plt.ylabel("Number of People")
plt.title("Distribution of Average Stream Duration", fontsize=20, pad=10)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show
```

```python
education_level = df["Education_Level"].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10,6))

plt.bar(education_level.index, education_level.values)

plt.show
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df['Age_at_Marriage'],
    df['Children_Count'],
    s=30
)

plt.xlabel('Age')
plt.ylabel('Number of Children')

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title("Does the Age at Marriage Impact the Number of Children?")

plt.tight_layout()
plt.show
```
